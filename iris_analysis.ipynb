```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv(r"C:\Users\kishl\OneDrive\Desktop\Iris.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.hist(figsize=(10,6), color='skyblue', edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot of Iris Dataset Features")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

num = df.select_dtypes(include=[np.number])    # keep only numeric columns
plt.figure(figsize=(8,6))
sns.heatmap(num.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (numeric cols only)")
plt.show()
```

```python
sns.pairplot(df, hue="Species")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(
    x="SepalLengthCm",
    y="PetalLengthCm",
    hue="Species",
    data=df
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["SepalLengthCm"], kde=True, color="purple")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Species", y="PetalWidthCm", data=df)
plt.title("Petal Width by Species")
plt.show()
```
